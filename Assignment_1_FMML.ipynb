```python
#Question 1
def reverse_and_concatenate(A, B):

    reversed_A = A[::-1]


    result = reversed_A + B

    return result


string_A = "Hello"
string_B = "World"
final_string = reverse_and_concatenate(string_A, string_B)
print(final_string)
```

```python
#Question 2
def find_most_hours_worked(work_hours):
    max_hours = 0
    female_worker = None

    for name, hours, gender in work_hours:
        if gender == 'F' and hours > max_hours:
            max_hours = hours
            female_worker = name

    if female_worker:
        payment = max_hours * 20
        print(f"The female worker that worked the most hours is {female_worker} and she should be paid {payment}")
    else:
        print("No female workers found in the list.")


work_hours = [('Abby', 100, 'F'), ('Billy', 400, 'M'), ('Cassie', 800, 'F'), ('Maggi', 600, 'F'),
              ('Alex', 500, 'M'), ('Raj', 225, 'M'), ('Penny', 920, 'F'), ('Ben', 300, 'M')]


find_most_hours_worked(work_hours)
```
